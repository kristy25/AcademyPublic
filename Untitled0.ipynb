```python
def acid_cat(x):
  if x<0.21:
    return 'A'
  elif x<0.3:
    return 'B'
  else:
    return 'C'
```

```python
df['citric acid'].max()
```

```python
acid_cat(df['citric acid'].max())
```

```python
df['new citric cat']=df['citric acid'].apply(lambda x: acid_cat(x))
df[['new citric cat','citric acid']].sample(10)
```

```python
mask4 = (((df['residual sugar']<=1.9) | (df['residual sugar'] >=2.3)) & (df['sulphates'] <0.7301) & (df['new citric cat']=='B'))
df[mask4]
```

```python
list_to_keep = ['A','B']
```

```python
# isin
#value_counts()
mask6=((df['sulphates'] <0.7301) & df['new citric cat'].isin(list_to_keep))
df[mask6]['new citric cat'].value_counts()
```

```python
# ~ is to take the complementary
mask6=((df['sulphates'] <0.7301) & (~df['new citric cat'].isin(list_to_keep)))
df[mask6]['new citric cat'].value_counts()
```

```python
mask6=((df['sulphates'] <0.7301) & (df['new citric cat']!='A'))
df[mask6]['new citric cat'].value_counts()
```
